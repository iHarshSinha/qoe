```python
import pandas as pd
import numpy as np

from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error

print("\n==============================")
print("STEP 30: Loading engineered dataset")
print("==============================")

df = pd.read_csv("../data/engineered_dataset.csv")

print("Dataset loaded")

print("\nShape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())


print("\n==============================")
print("STEP 31: Removing useless features")
print("==============================")

if "QoS_packet-loss" in df.columns:
    df = df.drop(columns=["QoS_packet-loss"])
    print("Dropped QoS_packet-loss (no variance)")


print("\nNew shape:", df.shape)


print("\n==============================")
print("STEP 32: Splitting features and target")
print("==============================")

X = df.drop(columns=["MOS"])
y = df["MOS"]

print("Feature matrix shape:", X.shape)
print("Target shape:", y.shape)

print("\n==============================")
print("STEP 33: Cross Validation Setup")
print("==============================")

cv = RepeatedKFold(
    n_splits=5,
    n_repeats=10,
    random_state=42
)

print("Cross validation: 5 folds × 10 repeats")

scoring = "neg_root_mean_squared_error"


print("\n==============================")
print("STEP 34: Training models")
print("==============================")

models = {

"Linear Regression": LinearRegression(),

"KNN": KNeighborsRegressor(),

"Random Forest": RandomForestRegressor(),

"Gradient Boosting": GradientBoostingRegressor(),

"SVR": SVR(),

"Neural Network": MLPRegressor(max_iter=2000)

}

results = {}

for name, model in models.items():

    print("\nTraining:", name)

    scores = cross_val_score(
        model,
        X,
        y,
        cv=cv,
        scoring=scoring
    )

    scores = -scores  # convert negative RMSE to positive

    mean_score = scores.mean()
    std_score = scores.std()

    results[name] = mean_score

    print("RMSE:", round(mean_score,3))
    print("STD :", round(std_score,3))


print("\n==============================")
print("STEP 35: Model comparison")
print("==============================")

for model, score in sorted(results.items(), key=lambda x: x[1]):
    print(model, "RMSE:", round(score,3))
```
