```python
import pandas as pd

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
import joblib
import os

print("\n==============================")
print("STEP 36: Loading engineered dataset")
print("==============================")

df = pd.read_csv("../data/engineered_dataset.csv")

df = df.drop(columns=["QoS_packet-loss"])

X = df.drop(columns=["MOS"])
y = df["MOS"]

print("Dataset loaded")
print("Shape:", df.shape)


print("\n==============================")
print("STEP 37: Cross validation")
print("==============================")

cv = RepeatedKFold(
    n_splits=5,
    n_repeats=10,
    random_state=42
)


print("\n==============================")
print("STEP 38: Random Forest tuning")
print("==============================")

rf_model = RandomForestRegressor()

rf_params = {

"n_estimators": [100,200,300],

"max_depth": [None,5,10,20],

"min_samples_split": [2,5,10],

"min_samples_leaf": [1,2,4]

}

rf_grid = GridSearchCV(
    rf_model,
    rf_params,
    cv=cv,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)

rf_grid.fit(X,y)

print("Best RF parameters:")
print(rf_grid.best_params_)

print("Best RF RMSE:")
print(-rf_grid.best_score_)


print("\n==============================")
print("STEP 39: Gradient Boosting tuning")
print("==============================")

gb_model = GradientBoostingRegressor()

gb_params = {

"n_estimators": [100,200,300],

"learning_rate": [0.01,0.05,0.1],

"max_depth": [3,5,7],

"subsample": [0.8,1.0]

}

gb_grid = GridSearchCV(
    gb_model,
    gb_params,
    cv=cv,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)

gb_grid.fit(X,y)

print("Best GB parameters:")
print(gb_grid.best_params_)

print("Best GB RMSE:")
print(-gb_grid.best_score_)

print("\n==============================")
print("STEP 40: Saving best models")
print("==============================")

os.makedirs("../models", exist_ok=True)

# Save Random Forest model
rf_best = rf_grid.best_estimator_

joblib.dump(
    rf_best,
    "../models/best_random_forest.pkl"
)

print("Random Forest model saved")


# Save Gradient Boosting model
gb_best = gb_grid.best_estimator_

joblib.dump(
    gb_best,
    "../models/best_gradient_boosting.pkl"
)

print("Gradient Boosting model saved")
```
