```python
import pandas as pd
import numpy as np

print("\n==============================")
print("STEP 1: Loading datasets")
print("==============================")

# Main dataset
main_df = pd.read_csv("../VLCQoSData.csv", sep=';')

# MOS dataset
mos_df = pd.read_csv("../data/MOSData.csv", sep=';')

# Individual ratings
ratings_df = pd.read_csv("../data/RatingsData.csv", sep=';')

# HRC configurations
hrc_df = pd.read_csv("../config/HRCsList.csv", sep=';')

print("\nDatasets loaded successfully.\n")

print("Main dataset shape:", main_df.shape)
print("MOS dataset shape:", mos_df.shape)
print("Ratings dataset shape:", ratings_df.shape)
print("HRC dataset shape:", hrc_df.shape)
```
