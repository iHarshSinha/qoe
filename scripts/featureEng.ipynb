```python
import pandas as pd
import numpy as np

print("\n==============================")
print("STEP 21: Loading dataset")
print("==============================")

df = pd.read_csv("../VLCQoSData.csv", sep=";")

print("Dataset loaded")

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nPreview:")
print(df.head())


print("\n==============================")
print("STEP 22: Cleaning packet loss column")
print("==============================")

print("Unique BEFORE cleaning:")
print(df["QoS_packet-loss"].unique())

df["QoS_packet-loss"] = df["QoS_packet-loss"].str.replace("%","").astype(float)

print("\nUnique AFTER cleaning:")
print(df["QoS_packet-loss"].unique())

print("\nColumn type:")
print(df["QoS_packet-loss"].dtype)


print("\n==============================")
print("STEP 23: Removing useless columns")
print("==============================")

drop_cols = ["id"]

df = df.drop(columns=drop_cols)

print("Columns removed:", drop_cols)

print("\nRemaining columns:")
print(df.columns.tolist())

print("\nNew shape:")
print(df.shape)


print("\n==============================")
print("STEP 24: Checking categorical columns")
print("==============================")

cat_cols = df.select_dtypes(include="object").columns

print("Categorical columns:")
print(cat_cols)

for col in cat_cols:
    print("\nColumn:", col)
    print("Unique values:", df[col].unique())


print("\n==============================")
print("STEP 25: Extracting screen resolution")
print("==============================")

df["screen_width"] = df["QoD_screen-resolotion"].str.split("x").str[0].astype(int)
df["screen_height"] = df["QoD_screen-resolotion"].str.split("x").str[1].astype(int)

df["screen_pixels"] = df["screen_width"] * df["screen_height"]

print("New columns created:")
print(["screen_width","screen_height","screen_pixels"])

print(df[["QoD_screen-resolotion","screen_pixels"]].head())


print("\n==============================")
print("STEP 26: Screen dimension processing")
print("==============================")

df["screen_dim_w"] = df["QoD_screen-dimension"].str.split("x").str[0].astype(int)
df["screen_dim_h"] = df["QoD_screen-dimension"].str.split("x").str[1].astype(int)

df["screen_area"] = df["screen_dim_w"] * df["screen_dim_h"]

print("New screen dimension features created")

print(df[["QoD_screen-dimension","screen_area"]].head())


print("\n==============================")
print("STEP 27: Dropping original text columns")
print("==============================")

drop_cols = [
"QoD_screen-resolotion",
"QoD_screen-dimension"
]

df = df.drop(columns=drop_cols)

print("Dropped columns:", drop_cols)

print("\nRemaining columns count:", df.shape[1])


print("\n==============================")
print("STEP 28: Encoding categorical variables")
print("==============================")

df = pd.get_dummies(df, drop_first=True)

print("Encoding completed")

print("\nNew dataset shape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())


print("\n==============================")
print("STEP 29: Saving engineered dataset")
print("==============================")

df.to_csv("../data/engineered_dataset.csv", index=False)

print("Engineered dataset saved at:")
print("../data/engineered_dataset.csv")

print("\nFinal dataset preview:")
print(df.head())

print("\nFinal dataset shape:")
print(df.shape)
```
